creamos la tabla en la base de datos

```python
from conexion_db import conectar_db


def create_table():
    conexion = conectar_db()
    cursor = conexion.cursor()

    cursor.execute("""
        CREATE TABLE IF NOT EXISTS grammy_awards (
            id SERIAL PRIMARY KEY,
            year INT,
            title TEXT,
            published_at TIMESTAMP,
            updated_at TIMESTAMP,
            category TEXT,
            nominee TEXT,
            artist TEXT,
            workers TEXT,
            img TEXT,
            winner BOOLEAN
        );
    """)

    conexion.commit()
    print("Tabla 'grammy_awards' creada exitosamente.")

    cursor.close()
    conexion.close()


if __name__ == "__main__":
    create_table()
```

cargamos los datos del csv a la db

```python
from conexion_db import conectar_db
import pandas as pd

def insertar_datos(df):
    conexion = conectar_db()
    cursor = conexion.cursor()

    for _, row in df.iterrows():
        cursor.execute("""
            INSERT INTO grammy_awards (year, title, published_at, updated_at, category, nominee, artist, workers, img, winner)
            VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
        """, (
            row['year'],
            row['title'],
            row['published_at'],
            row['updated_at'],
            row['category'],
            row['nominee'],
            row['artist'],
            row['workers'],
            row['img'],
            row['winner']
        ))

    conexion.commit()
    print("✅ Datos insertados correctamente.")

    cursor.close()
    conexion.close()

# Cargar CSV y llamar a la función
df = pd.read_csv("/home/isabella/Escritorio/Worksho_002/Grammy/the_grammy_awards.csv")  # Reemplaza con la ruta real
insertar_datos(df)
```

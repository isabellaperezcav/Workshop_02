This Jupyter Notebook performs an exploratory data analysis (EDA) on a Spotify dataset (spotify_dataset.csv). The dataset contains various musical features and metadata for tracks, such as danceability, energy, popularity, and genre. The notebook uses Python libraries (pandas, matplotlib, and seaborn) to load, inspect, and visualize the data.

## Importing Libraries 


We import the following libraries:

- `matplotlib.pyplot as plt` and `seaborn as sns` for data visualization.
- `pandas as pd` for data manipulation.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading Data 


We load the data from the "spotify_dataset.csv"  into a pandas DataFrame `df` using `pd.read_csv()`. The dataset contains 114000 rows and 21 columns

```python
df = pd.read_csv("spotify_dataset.csv")
df
```

# Visualization

Top 10 most popular songs 

```python
top_popular = df[['track_name', 'artists', 'popularity']].drop_duplicates().nlargest(10, 'popularity')

plt.figure(figsize=(12,6))
sns.barplot(x=top_popular['popularity'], y=top_popular['track_name'], palette="viridis")
plt.xlabel("Popularidad")
plt.ylabel("Canción")
plt.title("Top 10 Canciones Más Populares")
plt.show()
```

Top 10 music genres with more songs

```python
genre_counts = df['track_genre'].value_counts().head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=genre_counts.values, y=genre_counts.index, palette="coolwarm")
plt.xlabel("Cantidad de Canciones")
plt.ylabel("Género")
plt.title("Top 10 Géneros con Más Canciones")
plt.show()
```

how popular a song is according to its duration

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x=df['duration_ms'] / 60000, y=df['popularity'], alpha=0.6, color='green')
plt.xlabel("Duración (minutos)")
plt.ylabel("Popularidad")
plt.title("Relación entre Duración y Popularidad de Canciones")
plt.show()
```

Correlations between Musical Characteristics

```python
plt.figure(figsize=(12,6))
sns.heatmap(df[['danceability', 'energy', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'popularity']].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlations between Musical Characteristics")
plt.show()
```

- Strong Positive Correlations: energy and loudness (0.76) indicate that louder tracks tend to be more energetic.

- Moderate Positive Correlations: danceability and valence (0.49) suggest that happier tracks are often more danceable.
- Negative Correlations: acousticness and energy (-0.67) show that acoustic tracks tend to have lower energy.

- Weak Correlations with Popularity: popularity has low correlation with most features (e.g., 0.12 with danceability), suggesting that these musical attributes alone don’t strongly predict popularity.
